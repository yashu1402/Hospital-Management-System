```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
import mysql.connector
```

```python
win=Tk()
win.state('zoomed')
win.config(bg='white')

def pd():
    if e1.get()=="" or e2.get()=="":
        messagebox.showerror('Error','All Fields are required')
    else:
        con=mysql.connector.connect(host='localhost',username='root',password='12345',database='project')
        my_cursor=con.cursor()
        my_cursor.execute('insert into hospital values(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)',(
        nameoftablets.get(),ref.get(),dose.get(),nooftablets.get(),issd.get(),expd.get(),dd.get(),sideeffect.get(),nop.get(),dob.get(),pa.get()))
        con.commit()
        fetch_data()
        con.close()
        messagebox.showinfo("Success","Record has been inserted")

def fetch_data():
    con=mysql.connector.connect(host='localhost',username='root',password='12345',database='project')
    my_cursor=con.cursor()
    my_cursor.execute('select * from hospital')
    rows=my_cursor.fetchall()
    if len(rows)!=0:
        table.delete(* table.get_children())
        for items in rows:
            table.insert('',END,values=items)
        con.commit()
    con.close()

def get_data(event=''):
    cursor_row=table.focus()
    data=table.item(cursor_row)
    row=data['values']
    nameoftablets.set(row[0])
    ref.set(row[1])
    dose.set(row[2])
    nooftablets.set(row[3])
    issd.set(row[4])
    expd.set(row[5])
    dd.set(row[6])
    sideeffect.set(row[7])
    nop.set(row[8])
    dob.set(row[9])
    pa.set(row[10])

def pre():
    txt_frme.insert(END,'Name of Tablets:\t\t'+nameoftablets.get()+'\n')
    txt_frme.insert(END,'Reference No:\t\t'+ref.get()+'\n')
    txt_frme.insert(END,'Dose:\t\t'+dose.get()+'\n')
    txt_frme.insert(END,'Number of Tablets:\t\t'+nooftablets.get()+'\n')
    txt_frme.insert(END,'Issue Date:\t\t'+issd.get()+'\n')
    txt_frme.insert(END,'Expiry Date:\t\t'+expd.get()+'\n')
    txt_frme.insert(END,'Daily Dose:\t\t'+dd.get()+'\n')
    txt_frme.insert(END,'Side Effect:\t\t'+sideeffect.get()+'\n')
    txt_frme.insert(END,'Blood Pressure:\t\t'+bp.get()+'\n')
    txt_frme.insert(END,'Storage:\t\t'+storage.get()+'\n')
    txt_frme.insert(END,'Medication:\t\t'+med.get()+'\n')
    txt_frme.insert(END,'Name of Patients:\t\t'+nop.get()+'\n')
    txt_frme.insert(END,'D.O.B:\t\t'+dob.get()+'\n')
    txt_frme.insert(END,'Patient Address:\t\t'+pa.get()+'\n')
    
def delete():
    con=mysql.connector.connect(host='localhost',username='root',password='12345',database='project')
    my_cursor=con.cursor()
    querry=('delete from hospital where Ref_no=%s')
    value=(ref.get(),)
    my_cursor.execute(querry,value)
    con.commit()
    con.close()
    fetch_data()
    messagebox.showinfo('Deleted','Patient data has been deleted')
    
def clr():
    nameoftablets.set('')
    ref.set('')
    dose.set('')
    nooftablets.set('')
    issd.set('')
    expd.set('')
    dd.set('')
    sideeffect.set('')
    nop.set('')
    dob.set('')
    pa.set('')
    txt.frme.delete(1.0,END)

def exit():
    confirm=messagebox.askyesno('Confirmation','Are You Sure You Want To Exit')
    if confirm>0:
        win.destroy()
        return

Label(win,text="Hospital Management System",bd=10,font=('impack 25 bold'),bg='white',fg='Dark red').pack(fill=X)

frame1=Frame(win,bd=20,relief=RIDGE)
frame1.place(x=0,y=54,width=1540,height=350)

lf1=LabelFrame(frame1,text='Patient Information',font='arial 10 bold',bd=10,bg='white')
lf1.place(x=10,y=0,width=740,height=300)

Label(lf1,text='Name of Tablets',bg='white').place(x=10,y=10)
Label(lf1,text='Reference No',bg='white').place(x=10,y=40)
Label(lf1,text='Dose',bg='white').place(x=10,y=70)
Label(lf1,text='No. of Tablets',bg='white').place(x=10,y=100)
Label(lf1,text='Issue Date',bg='white').place(x=10,y=130)
Label(lf1,text='Expiry Date',bg='white').place(x=10,y=160)
Label(lf1,text='Daily Dose',bg='white').place(x=10,y=190)
Label(lf1,text='Side Effect',bg='white').place(x=10,y=220)
Label(lf1,text='Blood Pressure',bg='white').place(x=370,y=10)
Label(lf1,text='Storage ',bg='white').place(x=370,y=40)
Label(lf1,text='Medication',bg='white').place(x=370,y=70)
Label(lf1,text='Patient Id',bg='white').place(x=370,y=100)
Label(lf1,text='Name of Patient',bg='white').place(x=370,y=130)
Label(lf1,text='D.O.B',bg='white').place(x=370,y=160)
Label(lf1,text='Patient Address',bg='white').place(x=370,y=190)

nameoftablets=StringVar()
ref=StringVar()
dose=StringVar()
nooftablets=StringVar()
issd=StringVar()
expd=StringVar()
dd=StringVar()
sideeffect=StringVar()
bp=StringVar()
storage=StringVar()
med=StringVar()
pid=StringVar()
nop=StringVar()
dob=StringVar()
pa=StringVar()

e1=Entry(lf1,bd=4,textvariable=nameoftablets)
e1.place(x=130,y=10,width=200)

e2=Entry(lf1,bd=4,textvariable=ref)
e2.place(x=130,y=40,width=200)

e3=Entry(lf1,bd=4,textvariable=dose)
e3.place(x=130,y=70,width=200)

e4=Entry(lf1,bd=4,textvariable=nooftablets)
e4.place(x=130,y=100,width=200)

e5=Entry(lf1,bd=4,textvariable=issd)
e5.place(x=130,y=130,width=200)

e6=Entry(lf1,bd=4,textvariable=expd)
e6.place(x=130,y=160,width=200)

e7=Entry(lf1,bd=4,textvariable=dd)
e7.place(x=130,y=190,width=200)

e8=Entry(lf1,bd=4,textvariable=sideeffect)
e8.place(x=130,y=220,width=200)

e9=Entry(lf1,bd=4,textvariable=bp)
e9.place(x=480,y=10,width=200)

e10=Entry(lf1,bd=4,textvariable=storage)
e10.place(x=480,y=40,width=200)

e11=Entry(lf1,bd=4,textvariable=med)
e11.place(x=480,y=70,width=200)

e12=Entry(lf1,bd=4,textvariable=pid)
e12.place(x=480,y=100,width=200)

e13=Entry(lf1,bd=4,textvariable=nop)
e13.place(x=480,y=130,width=200)

e14=Entry(lf1,bd=4,textvariable=dob)
e14.place(x=480,y=160,width=200)

e15=Entry(lf1,bd=4,textvariable=pa)
e15.place(x=480,y=190,width=200)

lf2=LabelFrame(frame1,text='Prescription',font='arial 10 bold',bd=10,bg='white')
lf2.place(x=750,y=0,width=750,height=300)

txt_frme=Text(lf2,font='impack 10 bold',width=40,height=30)
txt_frme.pack(fill=BOTH)



frame2=Frame(win,bd=20,relief=RIDGE)
frame2.place(x=0,y=400,width=1540,height=330)

d_btn=Button(win,text='Delete',font=' ariel 20 bold',bg='white',fg='dark red',bd=6,cursor='hand2',command=delete)
d_btn.place(x=0,y=730,width=270)

p_btn=Button(win,text='Presciption',font=' ariel 20 bold',bg='white',fg='dark red',bd=6,cursor='hand2',command=pre)
p_btn.place(x=270,y=730,width=330)

pd_btn=Button(win,text='Prescription Data',font=' ariel 20 bold',bg='white',fg='dark red',bd=6,cursor='hand2',command=pd)
pd_btn.place(x=600,y=730,width=340)

c_btn=Button(win,text='Clear',font=' ariel 20 bold',bg='white',fg='dark red',bd=6,cursor='hand2',command=clr)
c_btn.place(x=940,y=730,width=330)

e_btn=Button(win,text='Exit',font=' ariel 20 bold',bg='white',fg='dark red',bd=6,cursor='hand2',command=exit)
e_btn.place(x=1270,y=730,width=300)

scroll_x=ttk.Scrollbar(frame2,orient=HORIZONTAL)
scroll_x.pack(side='bottom',fill='x')

scroll_y=ttk.Scrollbar(frame2,orient=VERTICAL)
scroll_y.pack(side='right',fill='y')

table=ttk.Treeview(frame2,columns=('not','ref','dose','nots','issd','expd','dd','sd','pn','dob','pa'),xscrollcommand=scroll_y.set,yscrollcommand=scroll_x.set)
scroll_x=ttk.Scrollbar(command=table.xview)
scroll_y=ttk.Scrollbar(command=table.yview)

table.heading('not',text="Name of Tablets")
table.heading('ref',text="Reference Number")
table.heading('dose',text="Dose")
table.heading('nots',text="No. of Tablets")
table.heading('issd',text="Issue Date")
table.heading('expd',text="Expiry Date")
table.heading('dd',text="Daily Dose")
table.heading('sd',text="Side Effects")
table.heading('pn',text="Patient Name")
table.heading('dob',text="D.O.B")
table.heading('pa',text="Patient Address")
table['show']='headings'
table.pack(fill=BOTH,expand=1)

table.column('not',width=100)
table.column('ref',width=100)
table.column('dose',width=100)
table.column('nots',width=100)
table.column('issd',width=100)
table.column('expd',width=100)
table.column('dd',width=100)
table.column('sd',width=100)
table.column('pn',width=100)
table.column('dob',width=100)
table.column('pa',width=100)
table.bind('<ButtonRelease-1>',get_data)
fetch_data()
mainloop()
```
